### Iris Data clustering

#### Importing Modules

```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans 
```

#### Loading Dataset

```python
iris_df = pd.read_csv('./datasets/Iris.csv',index_col = "Id")
iris = iris_df.drop("Species",axis = 1)
display(iris.head())

#shape of data

print('Shape of the Iris data : ',iris.shape)
```

```python
sns.pairplot(iris)
plt.show()
```

```python
pred = KMeans(n_clusters = 3,n_init  = 'auto').fit_predict(iris)
```

```python
plt_df = iris.copy(deep = True) 
plt_df['pred'] = pred
sns.pairplot(plt_df,hue  = "pred",palette = 'viridis')
```

#### Finding optimal value for k

```python
#finding optimal value for k

dist_scores=[]
K = np.arange(2,7)

for i in K:
    model = KMeans(n_clusters = i,n_init = 'auto',init = 'random')
    model.fit(iris)
    dist_scores.append(model.inertia_)

print(dist_scores)
```

```python
plt.plot(K,dist_scores,'s-',markersize = 5)
plt.xlabel('Number of clusters (K)')
plt.ylabel('Dist_score')
plt.title('Elbow Method for finding optimal k')
plt.show()
```

### Conclusion

From the Elbow method, It is known that there are 3 (optimal value of K) different clusters in the dataset
